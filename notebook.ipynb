```python
import requests
import json
```

```python
API_URL = "http://localhost:8000/predict"
print("Using API:", API_URL)
```

#Single request test

```python
payload = {"text": "This project turned out really well!"}

response = requests.post(API_URL, json=payload)

print("Input:", payload["text"])
print("Response:", response.json())
```

#Batch Test Multiple Requests

```python
texts = [
    "Mouse have unparalleld precision and accuracy when it comes to first person shooters",
    "Controllers are slippery and imprecise for first person shooter games",
    "This is the worst experience I've had",
    "This assignment was stressful but I made it :)"
]

for text in texts:
    r = requests.post(API_URL, json={"text": text})
    print(text, "→", r.json())
```
